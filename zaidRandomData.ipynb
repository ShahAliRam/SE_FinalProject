```python


sum_dict_morning={"monday_morning":0, 
         "tuesday_morning":0,
          "wednesday_morning":0, 
          "thursday_morning":0,
           "friday_morning":0,
          "saturday_morning":0,
           "sunday_morning":0
         }



sum_dict_evening={ "monday_evening":0, 
           "tuesday_evening":0,  
          "wednesday_evening":0,
          "thursday_evening":0,
          "friday_evening":0, 
        "saturday_evening":0, 
         "sunday_evening":0}




def change(df):
    days_list=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
    new_list_morning=[]
    new_list_evening=[]


    for i in days_list:
        new_list_morning.append(df.loc[i][0:9].sum())
        new_list_evening.append(df.loc[i][9:].sum())

    for i, key in enumerate(sum_dict_morning.keys()):
        sum_dict_morning[key] = new_list_morning[i]


    for i, key in enumerate(sum_dict_evening.keys()):
        sum_dict_evening[key] = new_list_evening[i]



    sum_dict_morning.update(sum_dict_evening)

    max_key = max(sum_dict_morning, key=sum_dict_morning.get)
    max_value = sum_dict_morning[max_key]


    min_value = min(sum_dict_morning, key=sum_dict_morning.get)
    min_key = sum_dict_morning[min_value]
    
    
    df_w['Shift_Time'] = df_w['Shift_Time'].replace(max_key, min_value)

    return df_w

```

```python
import random
import pandas as pd

# create a list of customer names
customers = ['John', 'Mary', 'Bob', 'Alice', 'Tom']

# create a list of food items
food_items = ['Grilled Salmon', 'Beef Tenderloin', 'Lobster Bisque', 'Truffle Fries', 'Roasted Beet Salad',
              'Margherita Pizza', 'Cheeseburger', 'Pad Thai', 'Chicken Alfredo', 'Sushi Roll']

# create an empty dataframe to store the recommendations
df__ = pd.DataFrame(columns=['customer', 'food_recommendation'])

# generate 30 random recommendations and append them to the dataframe
for i in range(30):
    customer = random.choice(customers)
    food_item = random.choice(food_items)
    likes_food = random.choice([True, False])
    if likes_food:
        sentence = f"{customer} recommends the {food_item}. They really love it!"
    else:
        sentence = f"{customer} eat {food_item}, but it was not  good at all ."
        
    df__ = df__.append({'customer': customer, 'food_recommendation': sentence}, ignore_index=True)


```

```python
sentiments=[]
for  i in df__['food_recommendation']:
    score = sid.polarity_scores(i)
    if score['compound'] > 0.1:
        sentiments.append('Positive')
    elif score['compound'] < -0.1:
        sentiments.append('Negative')
    else:
        sentiments.append('Neutral')
df__["sentiments"]=sentiments
```

```python
count_pos=0
count_neg=0
for i in df__["sentiments"]:
    if i =="Positive":
        count_pos=count_pos+1
    else:
        count_neg=count_neg+1
```

```python
total_pos_per=count_pos/len(df__["sentiments"])*100
total_pos_per
```

```python
total_neg_per=count_neg/len(df__["sentiments"])*100
total_neg_per
```
